```r
library(gridExtra)
library(tidyverse)
```

```r
smpl <- c(sample(c(0, 1), prob=c(0.3, 0.7), replace=TRUE, size=10))
```

```r
x <- replicate(100, {
    smpl <- c(sample(c(0, 1), prob=c(0.3, 0.7), replace=TRUE, size=25))
    mean(smpl)
})
```

```r
p1 <- data.frame(x_hat=x) %>% ggplot(aes(X_hat)) %>%
    + geom_histogram(binwidth=0.005, color="black")
p2 <- data.frame(x_hat=x) %>% ggplot(aes(X_hat)) %>%
    + stat_qq(dparams = list(mean=mean(X_hat), sd=sd(X_hat)))
    + geom_abline()
    + ylab("X_hat")
    + xlab("Theoritical normal")
grid.arrange(p1, p2, nrow=1)
```

```r
p <- seq(0, 1, length=100)
sample_sizes <- c(25, 100, 1000)
for (N in sample_sizes) {
    se <- sqrt(p * (1 - p) / N)
    plot(p, se, ylim=c(0, 0.10))
}
```
